## Micrograd

Micrograd is a small educational tool that demonstrates the core idea of automatic differentiation used in neural networks. It’s often used to understand how modern deep learning frameworks like PyTorch and TensorFlow implement backpropagation.

LETS get started ..


```python
from micrograd.engine import Value

x = Value(2.0)
y = Value(3.0)
z = x * y
z.backward()
w = z + x
print(f"Value of x: {x.data}, Gradient of x: {x.grad}")
print(f"Value of y: {y.data}, Gradient of y: {y.grad}")
print(f"Value of z: {z.data}, Gradient of z: {z.grad}")
print(f"Value of w: {w.data}, Gradient of w: {w.grad}")
```
